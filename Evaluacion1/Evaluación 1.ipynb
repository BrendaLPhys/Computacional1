```python
# Cargar a la memoria de trabajo las bibliotecas: Pandas (manejo de datos, Numpy (numerical python) y la biblioteca de gráficas Matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Usar "Shift+Enter" para procesar la información de la celda
```

```python
#Se leen los archivos correspondientes.

df=pd.read_csv('sargento_201117.csv', skiprows=2, skipfooter=1,engine='python', header=None, names=['#', 'Date', 'AbsPres', 'Temp', 'WaterLevels'])
df.head()
```

```python
df1 = pd.read_csv('sargento-salinidad-201117.csv', skiprows=3, header=None, names=['#', 'Date', 'CondHigh', 'Temp', 'SpecifConduct', 'Salinity'])
df1.head()
```

```python
# Convertir la cadena de caracteres 'Date' en variable temporal 'NDate'
df['Ndate'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %H:%M:%S')
df['month'] = df['Ndate'].dt.month
df.head()
```

```python
box1 = sns.boxplot(x="month", y="WaterLevels", data=df)
plt.show()
```

```python
# Convertir la cadena de caracteres 'Date' en variable temporal 'NDate'
df1['Ndate'] = pd.to_datetime(df1['Date'], format='%m/%d/%Y %H:%M:%S')
df1['month'] = df1['Ndate'].dt.month
df1.head()
```

```python
box2 = sns.boxplot(x="month", y="Salinity", data=df1)
plt.show()
```

```python
box3 = sns.boxplot(x="month", y="Temp", data=df1)
plt.show()
```
