```python
# Cargar a la memoria de trabajo las bibliotecas: Pandas (manejo de datos, 
# Numpy (numerical python) y la biblioteca de gr√°ficas Matplotlib
# Se asignan nombres cortos.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Lectura
df0=pd.read_csv('BudapestJun.txt', skiprows=3, sep='\s+')
# Dar estructura de datos (DataFrame)
df=pd.DataFrame(df0)
#Lectura
df1=pd.read_csv('BudapestDec.txt', skiprows=3, sep='\s+')
# Dar estructura de datos (DataFrame)
df2=pd.DataFrame(df1)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df2.describe()
```

```python
df2.dtypes
```

```python
# to change use .astype() 
df['HGHT'] = df.HGHT.astype(float)
df.dtypes
```

```python
df2['HGHT'] = df2.HGHT.astype(float)
df2.dtypes
```

```python
#La primer tarea es producir una gr√°fica de la variaci√≥n de la presi√≥n con respecto a la altura.
df.plot(x='PRES', y='HGHT')
plt.xlim(0,1000)
plt.title("Variaci√≥n de la presi√≥n VS altura (Junio)")
plt.ylabel("Altura (m)")
plt.xlabel("Presi√≥n (hPa)")
plt.grid(True)
plt.show()
```

```python
df2.plot(x='PRES', y='HGHT')
plt.xlim(0,1000)
plt.title("Variaci√≥n de la presi√≥n VS altura (Diciembre)")
plt.ylabel("Altura (m)")
plt.xlabel("Presi√≥n (hPa)")
plt.grid(True)
plt.show()
```

```python
#La siguiente parte consiste en hacer una gr√°fica de temperatura como funci√≥n de la altura.
df.plot(x='TEMP', y='HGHT')
plt.xlim(-80,30)
plt.title("Variaci√≥n de la Temperatura VS Altura (Junio)")
plt.ylabel("Altura (m)")
plt.xlabel("Temperatura")
plt.grid(True)
plt.show()
```

```python
df2.plot(x='TEMP', y='HGHT')
plt.title("Variaci√≥n de la Temperatura VS Altura (Diciembre)")
plt.xlim(-80,20)
plt.ylabel("Altura (m)")
plt.xlabel("Temperatura")
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la temperatura (TEMP) y temperatura de roc√≠o (DEWPT), como funci√≥n de la altura.
x1=df['TEMP']
x2=df['DWPT']
y=df['HGHT']

plt.plot(x1,y, "blue", label='TEMP')
plt.plot(x2,y, "orange", label='DWPT')
plt.title("Variaci√≥n de la Temperatura y Temperatura de roc√≠o VS Altura (Junio)")
plt.ylabel("Altura (m)")
plt.xlabel("Temperatura/Temperatura de roc√≠o")
plt.grid(True)
plt.show()
```

```python
x1=df2['TEMP']
x2=df2['DWPT']
y=df2['HGHT']

plt.plot(x1,y, "purple", label='TEMP')
plt.plot(x2,y, "black", label='DWPT')
plt.title("Variaci√≥n de la Temperatura y Temperatura de roc√≠o VS Altura (Diciembre)")
plt.ylabel("Altura (m)")
plt.xlabel("Temperatura/Temperatura de roc√≠o")
plt.grid(True)
plt.show()
```

```python
# to change use .astype() 
df['SKNT'] = df.SKNT.astype(float)
df.dtypes
```

```python
#Gr√°fica de la rapidez de los vientos en nudos (SKNT). 
plt.figure(); df.SKNT.plot()
plt.title("Rapidez de los Vientos (Junio)")
plt.ylabel("Rapidez (nudos)")
plt.grid(True)
plt.show()
```

```python
plt.figure(); df2.SKNT.plot()
plt.title("Rapidez de los Vientos (Diciembre)")
plt.ylabel("Rapidez (nudos)")
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la humedad relativa como funci√≥n de la altura.  
df.plot(x='RELH', y='HGHT')
plt.title("Variaci√≥n de la Humedad Relativa VS Altura (Junio)")
plt.xlim(0,70)
plt.ylabel("Altura (m)")
plt.xlabel("Humedad Relativa")
plt.grid(True)
plt.show()
```

```python
df2.plot(x='RELH', y='HGHT')
plt.title("Variaci√≥n de la Humedad Relativa VS Altura (Diciembre)")
plt.xlim(0,120)
plt.ylabel("Altura (m)")
plt.xlabel("Humedad Relativa")
plt.grid(True)
plt.show()
```
