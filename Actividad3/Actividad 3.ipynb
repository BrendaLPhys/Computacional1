```python
# Cargar a la memoria de trabajo las bibliotecas: Pandas (manejo de datos, 
# Numpy (numerical python) y la biblioteca de gráficas Matplotlib
# Se asignan nombres cortos.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Lectura
df0=pd.read_csv('BudapestJun.txt', skiprows=3, sep='\s+')
# Dar estructura de datos (DataFrame)
df=pd.DataFrame(df0)
#Lectura
df1=pd.read_csv('BudapestDec.txt', skiprows=3, sep='\s+')
# Dar estructura de datos (DataFrame)
df2=pd.DataFrame(df1)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df2.describe()
```

```python
df2.dtypes
```

```python
# to change use .astype() 
df['HGHT'] = df.HGHT.astype(float)
df.dtypes
```

```python
df2['HGHT'] = df2.HGHT.astype(float)
df2.dtypes
```

```python
#La primer tarea es producir una gráfica de la variación de la presión con respecto a la altura.
df.plot(x='PRES', y='HGHT')
plt.xlim(0,1000)
plt.title("Variación de la presión VS altura (Junio)")
plt.ylabel("Altura (m)")
plt.xlabel("Presión (hPa)")
plt.grid(True)
plt.show()
```

```python
df2.plot(x='PRES', y='HGHT')
plt.xlim(0,1000)
plt.title("Variación de la presión VS altura (Diciembre)")
plt.ylabel("Altura (m)")
plt.xlabel("Presión (hPa)")
plt.grid(True)
plt.show()
```

```python
#La siguiente parte consiste en hacer una gráfica de temperatura como función de la altura.
df.plot(x='TEMP', y='HGHT')
plt.xlim(-80,30)
plt.title("Variación de la Temperatura VS Altura (Junio)")
plt.ylabel("Altura (m)")
plt.xlabel("Temperatura")
plt.grid(True)
plt.show()
```

```python
df2.plot(x='TEMP', y='HGHT')
plt.title("Variación de la Temperatura VS Altura (Diciembre)")
plt.xlim(-80,20)
plt.ylabel("Altura (m)")
plt.xlabel("Temperatura")
plt.grid(True)
plt.show()
```

```python
#Gráfica de la temperatura (TEMP) y temperatura de rocío (DEWPT), como función de la altura.
x1=df['TEMP']
x2=df['DWPT']
y=df['HGHT']

plt.plot(x1,y, "blue", label='TEMP')
plt.plot(x2,y, "orange", label='DWPT')
plt.title("Variación de la Temperatura y Temperatura de rocío VS Altura (Junio)")
plt.ylabel("Altura (m)")
plt.xlabel("Temperatura/Temperatura de rocío")
plt.grid(True)
plt.show()
```

```python
x1=df2['TEMP']
x2=df2['DWPT']
y=df2['HGHT']

plt.plot(x1,y, "purple", label='TEMP')
plt.plot(x2,y, "black", label='DWPT')
plt.title("Variación de la Temperatura y Temperatura de rocío VS Altura (Diciembre)")
plt.ylabel("Altura (m)")
plt.xlabel("Temperatura/Temperatura de rocío")
plt.grid(True)
plt.show()
```

```python
# to change use .astype() 
df['SKNT'] = df.SKNT.astype(float)
df.dtypes
```

```python
df2['SKNT'] = df2.SKNT.astype(float)
df2.dtypes
```

```python
#Gráfica de la rapidez de los vientos en nudos (SKNT). 
plt.figure(); df.SKNT.plot()
plt.title("Rapidez de los Vientos (Junio)")
plt.ylabel("Rapidez (nudos)")
plt.grid(True)
plt.show()
```

```python
plt.figure(); df2.SKNT.plot()
plt.title("Rapidez de los Vientos (Diciembre)")
plt.ylabel("Rapidez (nudos)")
plt.grid(True)
plt.show()
```

```python
#Gráfica de la humedad relativa como función de la altura.  
df.plot(x='RELH', y='HGHT')
plt.title("Variación de la Humedad Relativa VS Altura (Junio)")
plt.xlim(0,70)
plt.ylabel("Altura (m)")
plt.xlabel("Humedad Relativa")
plt.grid(True)
plt.show()
```

```python
df2.plot(x='RELH', y='HGHT')
plt.title("Variación de la Humedad Relativa VS Altura (Diciembre)")
plt.xlim(0,120)
plt.ylabel("Altura (m)")
plt.xlabel("Humedad Relativa")
plt.grid(True)
plt.show()
```
